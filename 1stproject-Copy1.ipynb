```python
mp.figure(figsize=(7,4))

g = a.groupby("Region")[["Sales","Profit"]].sum()

mp.bar(g.index, g["Sales"])
mp.title("Sales by Region")
mp.xlabel("Region")
mp.ylabel("Sales")

mp.axhline(y=0, color='r', linestyle='--')

mp.show()
```

```python

grouped = a.groupby("Sub-Category")[["Sales","Profit"]].sum()

mp.figure(figsize=(10,3))
mp.bar(grouped.index, grouped["Profit"])
mp.xlabel("Products")
mp.xticks(rotation=60)
mp.ylabel("Profit")
mp.title("Loss Making Products")
mp.axhline(y=0, color="r", linestyle="--")
mp.show()
```
